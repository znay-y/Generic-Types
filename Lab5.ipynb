## Question 5 [2 marks] (must be explained to demonstrator)

A large estate agency has asked you to solve the following problem: They have a set of houses on sale, say $\{h_1, h_2, \ldots, h_n\}$, and also a set of clients $\{c_1, c_2, \ldots, c_m\}$. Each client wants to buy a house, and they have zero, one, or more houses that they prefer. Your job is to automatically match a house with a client, in such a way that the number of houses that are sold is maximised. 

For example, if we have only three clients $\{c_1, c_2, c_3\}$ and only three houses $\{h_1, h_2, h_3\}$, and $c_1$ would buy either of $\{h_1, h_2\}$, and $c_2$ would buy $h_1$, while $c_3$ would buy only $h_3$. We could depict this graphically as: 



Here you could match $c_1$ with $h_1$, and $c_3$ with $h_3$, but then you would have sold only two houses. A matching that maximises the number of houses sold is the one that matches $c_1\rightarrow h_2$, $c_2\rightarrow h_1$ and $c_3\rightarrow h_3$. 

Before you pull up your sleeves to start implementing a solution, you receive another request. This time from a hospital. They have a list of patients waiting for a transplant, and a list of organs ready for transplant. The hospital wants you to automatically match patients with compatible organs in such a way that you maximise the number of transplants. You realise that this is exactly the same problem as the one posed by the estate agents, so you decide to write generic Java code that can solve any such matching problem. 

To this end in the class `Match` you will implement the static method: 
```java
public static <U, T extends Compatibility<U>> Pair<T,U>[] match(T[] ts) 
```
Let us explain more the code that precedes it. First we have implemented a single method in the class `Tools`, namely 
```java
public static <T> T[] append(T[] ts, T t)
```
which simply takes an array of elements of type `T`, and one element of type `T` and appends it to the end of the array. Although it is better programming practice to use `List<T>`, we are using this version here since we have not introduced the `List` type yet. *You do not need to understand the implementation of this method, only how to use it*. 

Next we have the interface `Compatibility<U>`. Since in this matching problem we have to know whether an object is compatible with another one (e.g. is a client compatible with a house, meaning does the client want to buy it, or is an organ compatible with a patient), this interface gives the methods required for checking whether an object is compatible with another of type `U`. These methods are the self-evident `compatible` and `preferences` which returns an array of all objects that are compatible (in the client/house example, the preferences of a client will be an array of houses which the client would buy). 

We have given the implementations of `House` and `Client` classes. Note that the class `Client` implements the interface `Compatibility<House>`, meanning that we can compare clients for comaptibility with houses. The implementation is simple: the instance variable `wantsToBuy` remembers the preferences of a client. We also have the simple generic class `Pair` that holds a pair of objects. 

Finally we can go back to the method that you have to implement: 
```java
public static <U, T extends Compatibility<U>> Pair<T,U>[] match(T[] ts) 
```
It takes an array of type `T`, which is a type that implements the interface `Compatibility<U>`. So the objects of type `T` can be compared for compatibility with those of type `U`, and in turn returns a maximising matching in the form of an array of pairs. The example above is given in the test code below. 

*Hint:* Although there are complicated algorithm to solve this problem efficiently, do not implement them. Implement the brute force algorithm because it is the simplest (it tries all the possible matchings and chooses one that is maximal). 

**Write your answer below:**

````java
import java.util.Arrays;

class Tools {
    // Appends t to the (end of) array ts.
    // If array ts is null, create an array with one element and assign t to it
    public static <T> T[] append(T[] ts, T t) {
        if (ts == null) {
            @SuppressWarnings("unchecked")
            T[] result = (T[]) java.lang.reflect.Array.newInstance(t.getClass(), 1);
            result[0] = t;
            return result;
        }

        T[] result = Arrays.copyOf(ts, ts.length + 1);
        result[ts.length] = t;
        return result;
    }
}
````

````java
interface Compatibility<U> {
    public boolean compatible(U u);
    public U[] preferences();
}

class House {
    String name;
    public House(String name) { this.name = name; }
    public String toString() { return name; }
}
class Client implements Compatibility<House> {
    private House[] wantsToBuy = null;
    String name;

    public Client(String name) {
        this.name = name;
    }

    public String toString() {
        return name; 
    }

    // Add h to the list of houses this client would
    // like to buy. 
    public void likes(House h) {
        if (compatible(h)) return;
        wantsToBuy = Tools.append(wantsToBuy, h);
    }

    // Does this client want to buy the house h?
    public boolean compatible(House h) {
        if (wantsToBuy == null) return false; 
        for (House g : wantsToBuy) {
            if (g.equals(h)) return true;
        }
        return false;       
    }
    
    // Return the list of houses this client would 
    // like to buy.
    public House[] preferences() {
        return wantsToBuy;
    }
}

class Pair<T,U> {
    T left;
    U right; 

    public Pair(T left, U right) {
        this.left = left;
        this.right = right; 
    }
}

class Match {
    // Returns a maximal matching between objects of type T in the array ts
    // and their objects of preference of type U. 
    public static <U, T extends Compatibility<U>> Pair<T,U>[] match(T[] ts) {

    }
}
````

**Run your program:**


````java
House h1 = new House("h1");
House h2 = new House("h2");
House h3 = new House("h3");

Client c1 = new Client("c1");
Client c2 = new Client("c2");
Client c3 = new Client("c3");

c1.likes(h1);
c1.likes(h2);
c2.likes(h1); 
c3.likes(h3);

Client[] cs = {c1,c2,c3};
for(Pair<Client, House> p : Match.match(cs)) {
    System.out.print(p.left);
    System.out.print(" --- ");
    System.out.println(p.right);
}
````
